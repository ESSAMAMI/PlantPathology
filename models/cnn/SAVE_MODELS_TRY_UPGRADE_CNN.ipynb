```python
# model arnaud data aug qui vient de cifar10
# hyperparams
nb_layers = 8
kernel = (3,3)
my_optimizer = Adam(lr=0.001)
my_regularizer = regularizers.l1_l2(l1=0.0001 / nb_layers,
                                    l2= 0.0001 / nb_layers)

# model
model = Sequential()
#Input(shape=(img_size, img_size, 3))

#add aconvolutional group 1
model.add(Conv2D(64, kernel, activation='relu', 
                    kernel_regularizer=my_regularizer, padding='same',
                    input_shape=(img_size, img_size, 3), name='conv64_1'))
model.add(BatchNormalization(name='batchnorm_1'))
model.add(Conv2D(64, kernel, activation='relu', 
                    kernel_regularizer=my_regularizer, padding='same', name='conv64_2'))
model.add(BatchNormalization( name='batchnorm_2'))
model.add(AveragePooling2D((2, 2), padding='same', 
                                name='avg_pool_1'))
model.add(Dropout(0.4, 
    name='Dropout_1'))
          
#add aconvolutional group 2
model.add(Conv2D(128, kernel, activation='relu', 
                    kernel_regularizer=my_regularizer, padding='same', name='conv128_1'))
model.add(BatchNormalization( name='batchnorm_3'))
model.add(Conv2D(128, kernel, activation='relu', 
                    kernel_regularizer=my_regularizer, padding='same', name='conv128_2'))
model.add(BatchNormalization( name='batchnorm_4'))
model.add(AveragePooling2D((2, 2), padding='same', 
                                name='avg_pool_2'))
model.add(Dropout(0.4, 
    name='Dropout_2'))

#add aconvolutional group 3
model.add(Conv2D(256, kernel, activation='relu', 
                    kernel_regularizer=my_regularizer, padding='same', name='conv256_1'))
model.add(BatchNormalization( name='batchnorm_5'))
model.add(Conv2D(256, kernel, activation='relu', 
                    kernel_regularizer=my_regularizer, padding='same', name='conv256_2'))
model.add(BatchNormalization( name='batchnorm_6'))
model.add(AveragePooling2D((2, 2), padding='same', 
                                name='avg_pool_3'))
model.add(Dropout(0.4, 
    name='Dropout_3'))
          
#add aconvolutional group 4
model.add(Conv2D(512, kernel, activation='relu', 
                    kernel_regularizer=my_regularizer, padding='same', name='conv512_1'))
model.add(BatchNormalization( name='batchnorm_7'))
model.add(Conv2D(512, kernel, activation='relu', 
                    kernel_regularizer=my_regularizer, padding='same', name='conv512_2'))
model.add(BatchNormalization( name='batchnorm_8'))
model.add(AveragePooling2D((2, 2), padding='same', 
                                name='avg_pool_4'))
model.add(Dropout(0.4, 
    name='Dropout_4'))
          

#flatten and MLP
model.add(Flatten(name='Flatten'))
model.add(Dense(128, activation='relu', kernel_regularizer=my_regularizer,
                            name='Dense128_1'))
model.add(BatchNormalization( name='batchnorm_finale'))
model.add(Dropout(0.4, name='Dropout_final'))

# output layer
model.add(Dense(4, activation='softmax', name='output'))

model.compile(loss="categorical_crossentropy", optimizer=my_optimizer, metrics=["accuracy"])
```
